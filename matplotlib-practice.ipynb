Scatter


```python
import  matplotlib.pyplot  as plt

x=[3,7,1,8,3,9,10,43]
y=[2,66,22,8,1,99,3,55]
plt.title("scatter graph ")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.scatter(x,y,label="see scatter",color='r',marker='o')
plt.legend()
plt.grid(True,color='y')
plt.show()
```

```python
N_points = 100000
n_bins = 20

x = np.random.randn(N_points)
y = 4 * x + np.random.randn(100000) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
```

pi chart


```python

import  matplotlib.pyplot  as plt
from matplotlib import style
import numpy as np
style.use('ggplot')
slices=[1,2,6,10]
colors=['b','y','g','m']
days=[1,2,3,4,5]
activities=['travel','paint','photography','read']
plt.pie(slices,labels=activities,colors=colors,startangle=90, explode=(0.2,0.2,0,0.4))
plt.title("daily life")
plt.xlabel("x-axis")
plt.ylabel("y-label")
```

Stack graph

```python

import  matplotlib.pyplot  as plt
from matplotlib import style
import numpy as np

days=[1,2,3,4,5]
dance=[2,5,8,7,6]
sing=[2,5,7,1,8]
play=[5,8,12,9,1]
travel=[1,3,9,2,4]

plt.stackplot(days,dance,sing,play,travel,colors=['yellow','green','pink','blue'],labels=['dance','sing','play','travel'])
plt.title("routine")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend()
plt.show()
```
